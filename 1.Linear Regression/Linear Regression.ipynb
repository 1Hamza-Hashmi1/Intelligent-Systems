```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Load dataset from URL
url = "https://raw.githubusercontent.com/yoga-suhas-km/Intelligent_systems/main/marks_dataset.csv"
data = pd.read_csv(url)

# Data Preparation (EDA)
X = data['Midterm mark'].values
y = data['Final mark'].values

# Calculate mean and standard deviation for X and y
mean_X, std_X = np.mean(X), np.std(X)
mean_y, std_y = np.mean(y), np.std(y)

print(f"Midterm Marks - Mean: {mean_X}, Std Dev: {std_X}")
print(f"Final Marks - Mean: {mean_y}, Std Dev: {std_y}")

# Standardize X and y
X_standardized = (X - mean_X) / std_X
y_standardized = (y - mean_y) / std_y

# Define functions
def compute_error(m, b, X, y):
    N = len(y)
    error = np.sum((y - (m * X + b)) ** 2) / N
    return error

def gradient_descent(X, y, m, b, alpha, iterations):
    errors = []
    for _ in range(iterations):
        N = len(y)
        m_grad = -(2 / N) * np.sum(X * (y - (m * X + b)))
        b_grad = -(2 / N) * np.sum(y - (m * X + b))
        m -= alpha * m_grad
        b -= alpha * b_grad
        errors.append(compute_error(m, b, X, y))
    return m, b, errors

# Initial values
m_init = -0.5
b_init = 0
alpha = 0.0001
iterations_100 = 100
iterations_2000 = 2000

# Without Standardization
m_100, b_100, errors_100 = gradient_descent(X, y, m_init, b_init, alpha, iterations_100)
m_2000, b_2000, errors_2000 = gradient_descent(X, y, m_init, b_init, alpha, iterations_2000)

# With Standardization
m_std_100, b_std_100, errors_std_100 = gradient_descent(X_standardized, y_standardized, m_init, b_init, alpha, iterations_100)
m_std_2000, b_std_2000, errors_std_2000 = gradient_descent(X_standardized, y_standardized, m_init, b_init, alpha, iterations_2000)

# EDA Plot
plt.figure(figsize=(10, 6))
plt.scatter(X, y, color='blue', label='Data Points')
plt.xlabel("Midterm Marks")
plt.ylabel("Final Marks")
plt.title("Exploratory Data Analysis (EDA)")
plt.legend()
plt.show()

# Initial regression line
plt.figure(figsize=(10, 6))
plt.scatter(X, y, color='blue', label='Data Points')
plt.plot(X, m_init * X + b_init, color='red', label='Initial Regression Line (m=-0.5, b=0)')
plt.xlabel("Midterm Marks")
plt.ylabel("Final Marks")
plt.title("Initial Regression Line")
plt.legend()
plt.show()


# Regression line after 100 iterations (Without Standardization)
plt.figure(figsize=(10, 6))
plt.scatter(X, y, color='blue', label='Data Points')
plt.plot(X, m_100 * X + b_100, color='orange', label='Regression Line (100 iterations)')
plt.xlabel("Midterm Marks")
plt.ylabel("Final Marks")
plt.title("Regression Line After 100 Iterations (Without Standardization)")
plt.legend()
plt.show()

# Error Over 100 Iterations (Without Standardization)
plt.figure(figsize=(10, 6))
plt.plot(range(iterations_100), errors_100, label='Error (100 iterations)', color='orange')
plt.xlabel("Iteration")
plt.ylabel("Error")
plt.title("Error Over 100 Iterations (Without Standardization)")
plt.legend()
plt.show()

# Regression line after 2000 iterations (Without Standardization)
plt.figure(figsize=(10, 6))
plt.scatter(X, y, color='blue', label='Data Points')
plt.plot(X, m_2000 * X + b_2000, color='green', label='Regression Line (2000 iterations)')
plt.xlabel("Midterm Marks")
plt.ylabel("Final Marks")
plt.title("Regression Line After 2000 Iterations (Without Standardization)")
plt.legend()
plt.show()

# Error Over 2000 Iterations (Without Standardization)
plt.figure(figsize=(10, 6))
plt.plot(range(iterations_2000), errors_2000, label='Error (2000 iterations)', color='green')
plt.xlabel("Iteration")
plt.ylabel("Error")
plt.title("Error Over 2000 Iterations (Without Standardization)")
plt.legend()
plt.show()

# Regression line after 100 iterations (With Standardization)
plt.figure(figsize=(10, 6))
plt.scatter(X_standardized, y_standardized, color='blue', label='Data Points (Standardized)')
plt.plot(X_standardized, m_std_100 * X_standardized + b_std_100, color='red', label='Regression Line (100 iterations)')
plt.xlabel("Standardized Midterm Marks")
plt.ylabel("Standardized Final Marks")
plt.title("Regression Line After 100 Iterations (With Standardization)")
plt.legend()
plt.show()


# Regression line after 100 iterations (With Standardization)
plt.figure(figsize=(10, 6))
plt.scatter(X_standardized, y_standardized, color='blue', label='Data Points (Standardized)')
plt.plot(X_standardized, m_std_100 * X_standardized + b_std_100, color='red', label='Regression Line (100 iterations)')
plt.xlabel("Standardized Midterm Marks")
plt.ylabel("Standardized Final Marks")
plt.title("Regression Line After 100 Iterations (With Standardization)")
plt.legend()
plt.show()

# Regression line after 2000 iterations (With Standardization)
plt.figure(figsize=(10, 6))
plt.scatter(X_standardized, y_standardized, color='blue', label='Data Points (Standardized)')
plt.plot(X_standardized, m_std_2000 * X_standardized + b_std_2000, color='red', label='Regression Line (2000 iterations)')
plt.xlabel("Standardized Midterm Marks")
plt.ylabel("Standardized Final Marks")
plt.title("Regression Line After 2000 Iterations (With Standardization)")
plt.legend()
plt.show()


# Error Over 2000 Iterations (With Standardization)
plt.figure(figsize=(10, 6))
plt.plot(range(iterations_2000), errors_std_2000, label='Error (2000 iterations - Standardized)', color='purple')
plt.xlabel("Iteration")
plt.ylabel("Error")
plt.title("Error Over 2000 Iterations (With Standardization)")
plt.legend()
plt.show()
```

Ans:

When the learning rate was increased to 0.1, the gradient descent algorithm converged faster, producing a regression line that aligns well with the data points after 2000 iterations. In contrast, a smaller learning rate of 0.0001 caused the updates to the parameters to be too small, leading to an incorrect regression line with a negative slope, even after 2000 iterations. The error plots would further confirm this by showing minimal improvement for the smaller learning rate, while the higher learning rate achieves significant error reduction. This comparison highlights the importance of choosing an appropriate learning rate: too small, and the model converges too slowly or fails to converge; too high, and the model risks overshooting or diverging.

Q2: Investigate the effect of standardization.

Ans:

Standardization has a significant effect on the performance and behavior of gradient descent in linear regression. By transforming the data to have zero mean and unit variance, standardization ensures that all features are on the same scale, preventing features with larger magnitudes from dominating the gradient updates. Without standardization, the gradient descent algorithm may struggle to converge efficiently, as seen in slower or unstable error reduction and poorly aligned regression lines. In contrast, with standardization, the convergence becomes faster and more stable, as the gradients are more balanced and the algorithm can take consistent step sizes across all features. Additionally, standardization eliminates issues caused by large data magnitudes, such as numerical instability or overflow errors, making it particularly crucial for datasets with wide-ranging feature scales. Ultimately, standardization enhances the effectiveness and accuracy of the regression model, as demonstrated by the well-aligned regression line and smoother error reduction in standardized data.
